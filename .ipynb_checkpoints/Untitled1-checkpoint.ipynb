```python
# 表 1：2015 年 12 月之前的在线用户行为。（ijcai2016_taobao） 
"""
User_id:唯一用户id
Seller_id:唯一卖家id
Item_id:唯一商品id
Category_id:唯一类别 ID 
Online_Action_id:“0”表示“点击”，“1”表示“购买” 
Time_Stamp:日期
"""
df1 = pd.read_csv('ijcai2016_taobao.csv',header=1,usecols=['use_ID','sel_ID','ite_ID','cat_ID','act_ID','time'])
df1.use_ID = df1.use_ID.astype(str)
df1.ite_ID = df1.ite_ID.astype(str)
df1.act_ID = df1.act_ID.astype(str)
df1.sel_ID = df1.sel_ID.astype(str)
df1 = df1.rename(columns={'time\n':'time'})
df1.head()
```

```python
#  表 3：商家信息。 (ijcai2016_merchant_info) 
"""
Merchant_id:唯一商户id
Budget:强加于商家的预算限制
Location_id_list:可用位置列表，例如 1:356:89 
"""

df3 = pd.read_csv('ijcai2016_merchant_info.csv')
df3 = df3.rename(columns={'loc_list\n':'loc_list'})
df3.mer_ID = df3.mer_ID.astype(str)
df3.loc_list = df3.loc_list.astype(str)
df3.head()
```

```python
# 商家-位置
dd = pd.DataFrame()
df3 = df3.rename(columns={'loc_list\n':'loc_list'})
df3['loc_list'] = df3['loc_list'].str.strip('\n')
df3.mer_ID = df3.mer_ID.astype(str)
def func(line):
#     print(line)
    global dd
    loc_ls = line['loc_list'].split(':')
    if len(loc_ls) > 1:
        for i in range(1,len(loc_ls)):
            line2 = line
            line2['mer_ID']=str(line2['mer_ID'])
            line2['loc_list'] = loc_ls[i]
            dd = dd.append(line2)
    line3 = line
    line3['loc_list'] = loc_ls[0]
    return line3

df3 = df3.apply(func, axis=1).append(dd)
# df3.mer_ID = df3.mer_ID.astype(str)

df3
```

```python
def func(line):
#     print(line)
    n = len(line['act_ID'])
    line['len_act'] = n
    line['cut_act'] = line['act_ID'][-7:-2]  # 进五次购物行为为特征值
    line['target'] = line['act_ID'][-1] # 最后一次购物为目标值
    return line
records2 = records.reset_index().apply(func,axis=1)
records2
```
